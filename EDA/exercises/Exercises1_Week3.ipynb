```python
# Dataset
url='https://drive.google.com/file/d/1S4mHPFhdNYtIU0Agf_shFWFTLJTEqAWq/view?usp=sharing'
url='https://drive.google.com/uc?id=' + url.split('/')[-2]
```

```python
emp = pd.read_csv(url, parse_dates=['HIRE_DATE', 'JOB_DATE'])
```

```python
emp
```

```python
# Funcion para crear variable YEARS_EXPERIENCE usando 3 de octubre de 2022 como fecha actual
def yrs_exp(df_):
    days_hired = pd.to_datetime('10-3-2022') - df_.HIRE_DATE
    
    return days_hired.dt.days / 365.25
```

```python
emp = (emp.assign(YEARS_EXPERIENCE=yrs_exp))
```
